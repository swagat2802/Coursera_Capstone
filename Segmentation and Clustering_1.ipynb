```python
# import the libraries
import requests
import pandas as pd
import urllib.request
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = urllib.request.urlopen(url)
```

```python
# import the BeautifulSoup library so we can parse HTML and XML documents
from bs4 import BeautifulSoup
```

```python
# parse the HTML from our URL into the BeautifulSoup parse tree format
soup = BeautifulSoup(page, "lxml")
```

```python
# find all the table on the url
all_tables=soup.find_all("table")
all_tables
```

```python
# select the table which we want to analyze
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
# fetch the contents of each list and mearge
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
# convert into dataframe
df=pd.DataFrame(A,columns=['Postcode'])
df['Borough']=B
df['Neighbourhood']=C
df.head()
```

```python
# delete the rows where Borough = Not assigned
df = df[~df['Borough'].isin(['Not assigned'])]
```

```python
df.shape
```

```python
# get unique Borough values for Postcode
temp1 = df.groupby('Postcode')['Borough'].unique()
temp1
```

```python
# get unique Neighbourhood values for Postcode
temp2 = df.groupby('Postcode')['Neighbourhood'].unique()
temp2
```

```python
# combine both values
combined_df = pd.concat([temp1,temp2],axis = 1)
combined_df.reset_index(inplace = True)
combined_df.head()
```

```python
# apply lamda function to remove []
combined_df['Borough'] = combined_df['Borough'].apply(lambda x:','.join(map(str,x)))
combined_df['Neighbourhood'] = combined_df['Neighbourhood'].apply(lambda x:','.join(map(str,x)))
```
